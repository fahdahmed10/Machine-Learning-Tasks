```python
corr_metrix=df.corr().abs()
mask= np.triu(np.ones_like(corr_metrix, dtype=bool),k=1)
upper_triangle=corr_metrix.where(mask)
to_drop=[column for column in upper_triangle.columns if any(upper_triangle[column]>0.95)]
filterd_df=df.drop(columns=to_drop)
to_drop
```

```python
def boxplot_plot(data):
    plt.figure(figsize=(20, 20))
    i=1
    for column in filterd_df.columns:
        plt.subplot(16, 2, i)
        i+=1
        sns.boxplot(data=data, x=column, hue='diagnosis')
        plt.title(f'box_plot {column} by Diagnosis')
    plt.tight_layout()
    plt.show()     
boxplot_plot(filterd_df)
```

```python
def outlier_removal(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return series.where((series >= lower_bound) & (series <= upper_bound))
filterd_df = filterd_df.apply(outlier_removal)
filterd_df.dropna(inplace=True)
```

```python
boxplot_plot(filterd_df)
```

```python
filterd_df.to_csv('../data/cleaned_breast_cancer.csv', index=False)
```

```python
X = filterd_df.drop(columns=['diagnosis'])
Y = filterd_df['diagnosis']
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0, stratify=Y)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)   
```

```python
accuracy_table=[]
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
def results(model, x_train, y_train, x_test, y_test, accuracy_table):
    y_train_pred = model.predict(x_train)
    y_test_pred = model.predict(x_test)
    print("Train Accuracy:", accuracy_score(y_train, y_train_pred)*100,"%")
    print("Test Accuracy:", accuracy_score(y_test, y_test_pred)*100,"%")
    print("Classification Report:\n", classification_report(y_test, y_test_pred))
    cm = confusion_matrix( y_test, y_test_pred)
    accuracy_table.append({
        'Model': model.__class__.__name__,
        "Train Accuracy": accuracy_score(y_train, y_train_pred) * 100,
        'Test Accuracy': accuracy_score(y_test, y_test_pred) * 100
    })
    plt.figure(figsize=(8, 6))  
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(F'Confusion Matrix for {model.__class__.__name__}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')  
    plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
import joblib
model = LogisticRegression() 
model.fit(x_train, y_train)
results(model, x_train, y_train, x_test, y_test, accuracy_table)
joblib.dump(model, '../models/LogisticRegression.pkl')
```

```python
from sklearn.linear_model import SGDClassifier
model = SGDClassifier( loss='log_loss', 
    penalty='elasticnet',    
    alpha=0.05,            
    l1_ratio=0.3,           
    max_iter=1000,           
    tol=1e-3,                
    learning_rate='optimal',
    early_stopping=True,     
    class_weight='balanced',
    random_state=42)
model.fit(x_train, y_train)
results(model, x_train, y_train, x_test, y_test, accuracy_table)
joblib.dump(model, '../models/SGDClassifier.pkl')
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(x_train, y_train)
results(model, x_train, y_train, x_test, y_test, accuracy_table)
joblib.dump(model, '../models/SVC.pkl')
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=4,         
    weights='distance',    
    metric='euclidean',   
    p=2               
)
model.fit(x_train, y_train)
results(model, x_train, y_train, x_test, y_test, accuracy_table)
joblib.dump(model, '../models/KNeighborsClassifier.pkl')
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(
    criterion='gini',          
    max_depth=7,               
    min_samples_split=2,     
    min_samples_leaf=2,               
    random_state=42,
)
model.fit(x_train, y_train)
results(model, x_train, y_train, x_test, y_test, accuracy_table)
joblib.dump(model, '../models/DecisionTreeClassifier.pkl')
```

```python
from xgboost import XGBClassifier
model = XGBClassifier(
    n_estimators=100,        
    learning_rate=0.05,      
    max_depth=4,            
    subsample=.8,           
    colsample_bytree=.8,    
    gamma=1,                
    reg_alpha=0.1,  
    reg_lambda=1,   
    use_label_encoder=False, 
    eval_metric='logloss',
    random_state=42
)
model.fit(x_train, y_train)
results(model, x_train, y_train, x_test, y_test, accuracy_table)
joblib.dump(model, '../models/XGBClassifier.pkl')
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(  n_estimators=100,         
    max_depth=5,              
    min_samples_split=3,     
    min_samples_leaf=5,        
    max_features='log2',                  
    class_weight='balanced',   
    random_state=42)    
model.fit(x_train, y_train)
results(model, x_train, y_train, x_test, y_test, accuracy_table)
joblib.dump(model, '../models/RandomForestClassifier.pkl')
```

```python
accuracy_table = pd.DataFrame(accuracy_table)
accuracy_table.sort_values(by='Test Accuracy', ascending=False, inplace=True)
accuracy_table.to_csv('../models/accuracy_table.csv', index=False)
accuracy_table
```
